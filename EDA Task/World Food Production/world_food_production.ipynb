```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
world_food_production_df.groupby(['Entity', 'Year']).first()
```

```python
aggregations = {}
numerical_columns = world_food_production_df.select_dtypes(include=['number']).columns.drop(labels=['Year']).to_numpy()
for col in numerical_columns:
    aggregations[col] = 'sum'

yearly_production = world_food_production_df.groupby(['Year'], as_index=False).agg(aggregations)

yearly_production.head()
```

```python
num_rows = len(numerical_columns) // 2 + len(numerical_columns) % 2
fig, axes = plt.subplots(num_rows, 2, figsize=(20, 7 * num_rows))
for i, col in enumerate(numerical_columns):
    r = i // 2
    c = i % 2  # Corrected indexing here
    sns.lineplot(x='Year', y=yearly_production[col], data=yearly_production, ax=axes[r, c])  # Specify the axes parameter
    axes[r, c].set_ylabel("Production Per Year")
    axes[r, c].set_xlabel("Year")
    axes[r, c].tick_params(axis='x',rotation=30)
    axes[r, c].set_title(f"{col} Yearly Production")

if len(numerical_columns) % 2 != 0:
    fig.delaxes(axes[-1, -1])
fig.text(0.5, 0.9, "Yearly Production", va="center", ha="center", fontsize=14)
plt.show()
```

```python
afghanistan_production = world_food_production_df[world_food_production_df['Entity'].str.contains('Afghanistan')]
afghanistan_production
```

```python
num_rows = len(numerical_columns) // 2 + len(numerical_columns) % 2
fig, axes = plt.subplots(num_rows, 2, figsize=(20, 7 * num_rows))
for i, col in enumerate(numerical_columns):
    r = i // 2
    c = i % 2  # Corrected indexing here
    sns.lineplot(x='Year', y=afghanistan_production[col], data=yearly_production, ax=axes[r, c])  # Specify the axes parameter
    axes[r, c].set_ylabel("Production Per Year")
    axes[r, c].set_xlabel("Year")
    axes[r, c].tick_params(axis='x',rotation=30)
    axes[r, c].set_title(f"{col} Yearly Production")

if len(numerical_columns) % 2 != 0:
    fig.delaxes(axes[-1, -1])
fig.text(0.5, 0.9, "Afghanistan Yearly Production", va="center", ha="center", fontsize=14)
plt.show()
```
